We will arbitarily take window size 9. Initially I wanted 10 residues, but residues-1 (corrisponding to the residue of interest) must be even.

```python
def window_calc(myseq, residue_number, window_size):
    """
    
    """


    if window_size % 2 ==0: # even
        raise ValueError("window_size must be odd!")
    # if given a window frame of 9: -1 corresponds to the residue we want, which leaves 8/2= 4. 
    
    if window_size % 2 == 1: # odd
        n_width = int((window_size-1)/2)  # python requires it to be an int and not a float, for some reason.
        
        
        start =  residue_number - n_width
        end =  residue_number + n_width +1 # to include the actual residue
        


        # following code handles when the 0-th residue is called. it will give an empty str normally
        # this shifts the window to start counting from the right, instead
        if start<0:
            end += -start
            start = 0

        # the following code handles the reverse: if you cant count right and need to compensate left.
        if end> len(str(myseq)):
            start -= (end -len(str(myseq)))
            end = len(str(myseq))
        
        if residue_number>len(str(myseq)): # code if the residue number is unresonable
            raise ValueError("There aren't that many residues in the chain!")
        frame = myseq[start:end]
# unresonable window_size isn't considered, ie 23232323233 , as it will fail elsewhere, I hope.
    
    return str(frame)
```

```python
window_calc(myseq=myseq, residue_number=16, window_size=9)
```

```python
window_calc(myseq=myseq, residue_number=6232323233, window_size=9)
```
