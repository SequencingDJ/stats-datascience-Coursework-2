```python
from Bio import SeqIO
import pandas as pd

rows = []


for rec in SeqIO.parse(filename, "fasta"):
    parts = rec.description.split("|")

    # Extract PDB id (drop _1, lowercase)
    idcode = parts[0].split("_")[0].lower()

    # Extract chain letter from "Chain A"
    chain = parts[1].replace("Chain", "").strip()

    sequence = str(rec.seq)

    rows.append({
        "idcode": idcode,
        "chain": chain,
        "sequence": sequence
    })

df_fasta = pd.DataFrame(rows)
```

```python
import pandas as pd
import re

rows = []

for _, row in df_fasta.iterrows():
    # extract chain letters (single capital letters)
    chains = re.findall(r'\b[A-Z]\b', row["chain"])

    for ch in chains:
        rows.append({
            "idcode": row["idcode"],
            "chain": ch,
            "sequence": row["sequence"]
        })

df_clean = pd.DataFrame(rows)

df_clean[10:20]
```
